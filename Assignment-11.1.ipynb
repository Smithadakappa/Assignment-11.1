```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
titanic.head()
```

```python
titanic["sex"]
```

```python
import matplotlib.pyplot as plt
gender_data = titanic["sex"].value_counts()
colors = ["#1f77b4", "#ff7f0e"]
explode = (0.1,0)
plt.pie(gender_data, labels=gender_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("proportion of gender")
plt.show()
```

```python
import numpy as np 
titanic['sexnum'] = np.where(titanic['sex']=='male', 1, 0) 
```

```python
import matplotlib as mpl
scatter_plot = plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, 
                           c=titanic['sexnum'])
plt.show()
```
